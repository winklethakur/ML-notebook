```python
a = np.random.randint(1,10,6)
a
```

```python
# find max of this array.
np.max(a)
```

```python
# find min
np.min(a)
```

```python
# find the index of max value in an array.
np.argmax(a)
```

```python
# find the index of min value in an array.
np.argmin(a)
```

```python
a = np.random.randint(1,10,6)
a
```

```python
# change the odd value to -1 in an array.
a[a%2==1]=-1
a
```

```python
# change values using where clause.
a = np.random.randint(1,20,6)
a
```

```python
# where(condition,true,false)
out = np.where(a%2==1,-1,a)
out
```
