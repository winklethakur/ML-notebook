### Indexing using boolean arrays

```python
arr = np.random.randint(low=1,high=100,size=20).reshape(5,4)  # random values from 1 to 100.
arr
```

```python
# fetch values greater than 50.
```

```python
arr > 50
```

```python
arr[arr > 50]
```

```python
# fetch values greater than 50 and odd values.
```

```python
arr[(arr>50)&(arr%2!=0)]
```

```python
# Set all values which is odd and greater than 50 will be equal to zero:
```

```python
arr[(arr>50)&(arr%2!=0)] = 0
```

```python
arr
```

### Broadcasting

```python
a1 = np.arange(8).reshape(2,4)
a2 = np.arange(8,16).reshape(2,4)
print(a1)
print('--'*8)
print(a2)
print('final answer is :')
print(a1+a2)
```

```python
a3 = np.arange(4).reshape(4,1)     # in these cases, check adjacent value of 1 is same as the other array row or column.
a4 = np.arange(12).reshape(4,3) 
print(a3)
print(' '*8)
print(a4)
print(' '*8)
print(a3+a4)
```

```python
a5 = np.arange(3).reshape(1,3)
a6 = np.arange(16).reshape(4,4)    # throw error because no common row or column in both arrays
print(a5+a6)
```

```python
a7 = np.arange(3).reshape(1,3)
a8 = np.arange(3).reshape(3,1) # both arrays have 1 and corsponding value of 1 is same in both arrays so no error.
print(a7)
print(a8)
print(a7+a8)
```

```python
a9= np.arange(1).reshape(1,1)
a10= np.arange(20).reshape(4,5)
print(a9)  # a9 will broadcast itself into a10 shape and we will get the result.
print(a10)
```
