```python
import pandas as pd
import numpy as np
```

```python
date = pd.read_csv('orders.csv')
time = pd.read_csv('messages.csv')
```

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
time['date'] = pd.to_datetime(time['date'])
```

## 1. Extract year

```python
date['year'] = date['date'].dt.year
date.head()
```

## 2. Extract Month

```python
date['month'] = date['date'].dt.month
```

```python
date.head()
```

```python
# month name:
date['month_name'] = date['date'].dt.month_name()
```

```python
date.head()
```

## 3. Extract day

```python
date['day'] = date['date'].dt.day
date.head()
```

```python
# day of week :
date['week'] = date['date'].dt.dayofweek
date.head()
```

```python
# name of day
date['date_name'] = date['date'].dt.day_name()
date.head()
```

```python
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend?
date['weekend'] = np.where(date['date_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['product_id','city_id','orders']).head()
```

## Extract week of the year 

```python
date['week_of_year'] = date['date'].dt.isocalendar().week
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# extract quarter
date['quarter'] = date['date'].dt.quarter
date.drop(columns=['product_id','city_id','orders']).head()
```

### Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
date.drop(columns=['product_id','city_id','orders']).head()
```
