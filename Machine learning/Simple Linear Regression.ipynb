```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/winkle/Downloads/placement (1).csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('cgpa')
plt.ylabel('package')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import accuracy_score
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[1].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.plot(X_test.to_numpy().ravel(), lr.predict(X_test), color='red')
plt.xlabel('cgpa')
plt.ylabel('package')
plt.show()
```
