A column where numerical and categorical data both present.

To handle this kind of data in a column we need to split data 

```python
# eg:- In titanic data , there is a column as cabin have values -> B2, C35, H5 ,etc. 
# in this case we need to split the numerical and categorical data in 2 columns.
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot(kind='bar')
fig.set_title('Passengers travelling with')
```

```python
df['number_numerical'] = pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # captures numerical part
df['cabin_cat'] = df['Cabin'].str[0]  # captures the first letter
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
df['Ticket'].unique()
```

```python
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')

df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,df['ticket_cat'])
```
