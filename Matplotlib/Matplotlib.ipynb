```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# Plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma VS Virat Kohli Career comparison')
plt.xlabel('season')
plt.ylabel('Runs Scored')
```

```python
# change colors(hex):
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#F283E3')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#CFF283')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('season')
plt.ylabel('Runs Scored')
```

```python
# Line(width and style) 
plt.plot(batsman['index'],batsman['V Kohli'],color='#81438F',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ED220C',linestyle='dotted')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Dash dotted style line:
plt.plot(batsman['index'],batsman['V Kohli'],color='#ED220C',linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#81438F',linestyle='dashdot')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('season')
plt.ylabel('Runs Scored')
```

```python
# marker(size):
plt.plot(batsman['index'],batsman['V Kohli'],color='#ED220C',linestyle='dashdot',marker='D',markersize=5)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#81438F',linestyle='dashdot',marker='o',markersize=10)
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#ED220C',linestyle='dashdot',marker='D',markersize=5,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#81438F',linestyle='dashdot',marker='o',markersize=10,label='Rohit')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('season')
plt.ylabel('Runs Scored')

plt.legend()
```

```python
# limiting axes
```

```python
# grid:
plt.plot(batsman['index'],batsman['V Kohli'],color='#ED220C',linestyle='dashdot',marker='D',markersize=5,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#81438F',linestyle='dashdot',marker='o',markersize=10,label='Rohit')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('season')
plt.ylabel('Runs Scored')

plt.legend()
plt.grid()
```

```python
# show() :
# if the graph is not showing we use this function.
plt.plot(batsman['index'],batsman['V Kohli'],color='#ED220C',linestyle='dashdot',marker='D',markersize=5,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#81438F',linestyle='dashdot',marker='o',markersize=10,label='Rohit')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('season')
plt.ylabel('Runs Scored')

plt.legend()
plt.grid()
plt.show()
```

## Scatter Plots :
- Bivariate Analysis
- numerical vs numerical
- use case - Finding correlation

```python
# plt.scatter on pandas data
df = pd.read_csv('batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color = '#81438F')
plt.title('Avg and SR Analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('Srike rate')
```

```python
# Size
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)      # slower
```

```python
# size will show how many people came together.
```

```python
# Scatter plot using plt.plot
plt.plot(tips['total_bill'],tips['tip'],'o')   # faster
```

## Bar chart
- Bivariate and univariate
- Numerical VS Categorical
- Use case - Aggregate analysis of groups

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
#multiple bar chart:
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
# plt.xticks can replace the label 
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue blue','green green green green green green','yellow yellow yellow yellow','pink']
plt.bar(colors,children)
```

```python
# x labels are overlapping so we use xticks can rotate the lable vertical.
```

```python
plt.bar(colors,children)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# it will increase readability.
```

```python
# Stacked Bar Chart :
df
```

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016']+df['2017'],label='2015')
plt.legend()
```

## Pie Chart 
- Univariate/Bivariate Analysis
- Categorical vs Numerical
- Use case - To find a contribution on a standard scale

```python
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')  # autopct is used to show %age.
plt.show()
```

```python
# explode:
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)  
plt.show()
```
