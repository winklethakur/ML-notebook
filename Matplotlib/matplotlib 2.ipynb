```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('iris (2).csv')
```

```python
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
# Coloured Scattered :
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap = 'summer',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
# size change of graph:
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

## Annotations :

```python
df = pd.read_csv('batter.csv')
df
```

```python
sample_df = df.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

## 3D Scatter Plot :

```python
batters = pd.read_csv('batter.csv')
batters.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(18,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(batters['runs'], batters['avg'], batters['strike_rate'])
ax.set_xlabel('Runs')
ax.set_ylabel('Average')
ax.set_zlabel('Strike Rate')

plt.show()
```

```python
fig, ax= plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))
ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0].set_title('avg vs strike rate')
ax[0].set_ylabel('strike rate')

ax[1].scatter(batters['avg'],batters['runs'])
ax[1].set_title('avg vs runs')
ax[1].set_ylabel('runs')
```

```python
# plot 4 graphs:
fig, ax= plt.subplots(nrows=2,ncols=2,figsize=(10,6))
ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['runs'],color='green')
ax[1,1].hist(batters['avg'],color='pink')
```

```python
fig = plt.figure(figsize=(12,12))

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(batters['avg'],batters['runs'],color='black')

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])
```
