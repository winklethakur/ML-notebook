# Inheritance 

When one class(child/derived) derives(lena) the properties & methods of another class(parent/base).

 ### 1. Single Inheritance

```python
class Car:
    color = 'black'
    @staticmethod
    def start():
        print('car started..')
    @staticmethod
    def stop():
        print('car stopped.')
        
class ToyotaCar(Car):
    def __init__(self,brand):
        self.brand = brand
        
car1 = ToyotaCar('fortuner')
car2 = ToyotaCar('prius')
```

```python
print(car1.brand)
```

```python
print(car1.brand,car1.color)
```

```python
print(car1.start())
```

```python
print(car2.brand,car2.stop())
```

```python
class fortuner(ToyotaCar):
    def __init__(self,cartype):
        self.cartype = cartype
c1 = fortuner('diesel')
```

### 3. Multiple Inheritance

```python
'''
 Base1            Base2
(parent1)       (parent2)
   .             .
     .         .
       .     .
        Child
'''
```

```python
class A:
    varA = "welcome to class A"
    
class B:
    varB = 'welcome to class B'
    
class C(A,B):
    varC = 'welcome to class C'
    
c1 = C()
```

```python
print(c1.varA)
print(c1.varB)
print(c1.varC)
```

## Super method

```python
# super() method is used to access methods of the parent class.
```

```python
class Car:
    def __init__(self, Cartype):
        self.Cartype = Cartype
        
    @staticmethod
    def start():
        print('car started..')
        
    @staticmethod
    def stop():
        print('car stopped')
        
class ToyotaCar(Car):
    def __init__(self,name,Cartype):
        self.name=name
        super().__init__(Cartype)
        super().start()
        
car1 = ToyotaCar('prius','electric')
print(car1.Cartype)
```

```python
# Staticmethod is used on those method
# where there is no attribute present because it can not be accessed or modified
```
