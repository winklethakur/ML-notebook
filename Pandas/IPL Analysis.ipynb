### 1. Dropping missing data using dropna( ) :

```python
train.shape
```

```python
train.dropna().shape
```

```python
# by default axis is 0(row). if axis = 1 (column) and it will remove null values from column.
```

```python
# how = "any"(by default) which means if any value is null in a row it will drop that row. 
#and if how ='all' means all values are null in a row then only it will drop that row.
```

```python
train.dropna(how = 'all').shape
```

```python
train.dropna(axis=1).shape
```

```python
# subset = None (by default)
# we can add a particular column or multiple column as a list in subset 
# to remove null values from those columns only.
```

```python
train.dropna(subset='Cabin').shape
```

```python
train.dropna(subset=['Cabin','Embarked']).shape
```
