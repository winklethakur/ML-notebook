```python
def run_scored(batsman_name):
    vk = delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1).index[0]
```

### isin( )

```python
# isin()
# Find the most distructive death over batsman in the history of IPL
# Strike Rate = (Number of runs/Number of balls)*100
# Min batsman 200 balls in over 16-20 
```

```python
death_over = delivery[delivery['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x = all_batsman>200
```

```python
# boolean indexing 
all_batsman[x]
```

```python
all_batsman[x].index # shows the batsman column
```

```python
# convert index into a list using tolist():
batsman_list = all_batsman[x].index.tolist()
```

```python
# no. of runs :
# runs scored by all these 43 batsman
final = death_over[death_over['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
# Balls played by these 43 batsman:
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
# strike rate: 
sr = (runs/balls)*100
```

```python
sr
```

### Merge function:

```python
# Find the list of orange cap holders 
```

```python
# orange cap holders is the player who has scored highest run in that particular IPL season. 
```

```python
# Merge is same as join in sql.
```

```python
# so we need to add matches dataset also.
```

```python
match = pd.read_csv('matches - matches.csv')
```

```python
match.head()
```

```python
delivery.head()
```

```python
# (right)id and (left)match_id are the same columns in both the tables so we will join both with it only.
new=delivery.merge(match,left_on='match_id',right_on='id')
```

```python
new.shape
```

```python
print(delivery.shape)
print(match.shape)
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False)
```

```python
type(new.groupby(['season','batsman'])['batsman_runs'].sum())
```

```python
#now we will convert series into dataframe by using "reset_index()"
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index()
```

```python
# now we get highest run by batsman of every season,now we will take only highest run player and remove others.
# for that we will use drop_duplicates().
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first')
```

```python
# again sort in terms of season and take 2 columns only : season and batsman.
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')[['season','batsman']]
```

### Pivot table :
pivot_table()

```python
# find no of six in each over for each batting team using pivot table.
```

```python
mark = delivery['batsman_runs']==6
six = delivery[mark]
six.head()
```

```python
pt=six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count') # count no.of six in each over.
```

```python
pt
```

```python
# convert it into a heatmap.
```

```python
import seaborn as sns
```

```python
sns.heatmap(pt)
```
