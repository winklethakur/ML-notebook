```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
titanic = pd.read_csv('train.csv')
titanic.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

### 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker',size='size')   # multivariate analysis
```

### 2. Bar Plot (Numerical - Categorical)

```python
sns.barplot(data=titanic,x='Pclass',y='Age',hue='Sex')
```

### 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(data = titanic,x='Sex',y='Age',hue='Survived')
```

### 4. Distplot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.kdeplot(titanic[titanic['Survived']==0]['Age'])
sns.kdeplot(titanic[titanic['Survived']==1]['Age'])
```

```python
sns.kdeplot(data=titanic,x='Age',hue='Survived')
```

```python
sns.displot(data=titanic,x='Age',kind='kde',hue='Survived')
```

### 4. Heatmap (Categorical - Categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
flights.pivot(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(flights.pivot(index='month',columns='year',values='passengers'))
```

### 6. Pairplot

```python
sns.pairplot(iris)
```

### 7. Lineplot (Numerical - Numerical)

```python
flights.head()
```

```python
new = flights.groupby('year').sum().reset_index()
```

```python
sns.lineplot(data = new,x='year',y='passengers')
```
