```python
# write lines 
L = ['hi\n','how are you?\n',"I'm fine"]
f = open('sample.txt','w')
f.writelines(L)
f.close()
```

```python
# reading from file:
# -> Using read()
f = open('sample.txt','r')
s = f.read()
print(s)
f.close()
```

```python
# -> using readline()
f = open('sample.txt','r')
s = f.readline()
k = f.readline()
print(s)
print(k)
f.close()
```

```python
f = open('sample.txt','r')

while True:
    data=f.readline()
    if data=="":
        break
    else:
        print(data,end='')
f.close()
```

```python
with open('big.txt','r') as f:
    
    chunk_size = 50
    
    while len(f.read(chunk_size))>0:
        print(f.read(chunk_size))
        f.read(chunk_size)
```

```python
# seek and tell function:
with open('sample.txt','r') as f:
    f.seek(2)   # seek is used to tell where to start
    print(f.read(20))
    print(f.tell())  # tell is used to tell us how many text is processed
```

```python
with open('sample.txt','r') as f:
    f.seek(0)   
    print(f.read(20))
    print(f.tell()) 
    f.seek(10)
    print(f.read(15))
```

```python
# seek during write:
with open('sample1.txt','w') as f:
    f.write('winkle')
    f.seek(0)
    f.write('s')
```

### Serialization and Deserialization:
- **Serialization** - process of converting python data types to JSON format.
- **Deserialization** - process of converting JSON format to python data types

### What is JSON?
JSON (**java script on notation**), it is a universal text format understood by all other pragramming languages

```python
# serialization using json format
# list 
import json
L = [1,2,3,4,5]

with open('demo.json','w') as f:
    json.dump(L,f)
```

```python
d = {
    'name':'winkle',
    'age':24,
    'gender':'female'
}
```

```python
with open('demo.json','w') as f:
    json.dump(d,f,indent=4)
```

```python
# deserialization using json :
import json
with open('demo.json','r') as f:
    w=json.load(f)
    print(w)
    print(type(w))
```

```python
# serialization and deserialization can handle all data types.
```
