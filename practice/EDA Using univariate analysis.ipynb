```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

### 1. Categorical Data

#### a. Countplot

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(df['Embarked'])
```

#### b. PieChart

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

### 2. Numerical Data

#### a. Histogram

```python
sns.histplot(df['Age'])
```

```python
plt.hist(df['Age'])
```

#### b. Distplot

```python
sns.distplot(df['Age'])
```

```python
sns.displot(df['Age'],kind='kde')
```

```python
sns.histplot(df['Age'],kde=True)
```

#### c. Boxplot

```python
sns.boxplot(df['Age'])
```
